חיפוש העיר עם הכי הרבה והכי מעט תובים

```python
max_city = df['Population'].idxmax()
value_max = df.loc[max_city, ' "City"']
print(value_max)
min_city = df['Population'].idxmin()
value_min = df.loc[min_city, ' "City"']
print(value_min)
```
