```python
#ex1
months = ['ינואר', 'פברואר', 'מרץ', 'אפריל', 'מאי', 'יוני',
          'יולי', 'אוגוסט', 'ספטמבר', 'אוקטובר', 'נובמבר', 'דצמבר']
temperatures = [12, 13, 16, 20, 24, 27, 29, 28, 26, 22, 17, 13]
```

```python
plt.plot(months, temperatures, marker='o', color='blue', linestyle='-')
plt.title('שינוי בטמפרטורה לאורך השנה')
plt.xlabel('חודש')
plt.ylabel('טמפרטורה ממוצעת (°C)')
plt.grid(True)


plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```
